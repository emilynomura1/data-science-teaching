```python
# Import gapminder data
url = "https://drive.google.com/file/d/1FisnRZ-X0r6Gltezy1OSeZzVJ9Q1HTFm/view?usp=sharing"
url_clean = 'https://drive.google.com/uc?id=' + url.split('/')[-2]
gapminder = pd.read_csv(url_clean)
gapminder.head()
```

```python
# Your code here:
gapminder.dropna(inplace=True) # Remove NAs

# Plot single & paired distributions of each numeric variable
sns.pairplot(data=gapminder, hue='continent', plot_kws=dict(alpha=0.3))
plt.show()
```

```python
# Top 10 countries with highest GDP
gapminder.groupby("country").max("gdp").sort_values("gdp", ascending=False)[0:10]
```
