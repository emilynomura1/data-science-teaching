## Part 2 - Bootstrapping

Consider a simple linear regression model from simulated random normal data. 

```python
import statsmodels.api as sapi

n = 100
beta_0 = 2
beta_1 = 3
sigma = 2

x = nr.normal(0, 1, n)
x = np.array(sorted(x))
y = beta_0 + x * beta_1
y = y + np.random.uniform(low=-1, high=1, size=n) * sigma
df_dict = {'x': x, 'y': y}
df = pd.DataFrame(df_dict)

sns.scatterplot(x=x, y=y)
plt.show()
```
